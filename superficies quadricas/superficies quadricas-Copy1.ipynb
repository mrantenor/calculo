# DERIVADAS PARCIAIS


```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
from numpy import sin, cos, sinh, cosh, pi, sqrt
```

```python
class eq_par:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z
        
class eq_par_der:
    def __init__(self, dx_du, dy_du, dz_du, dx_dv, dy_dv, dz_dv):
        self.dx_du = dx_du
        self.dy_du = dy_du
        self.dz_du = dz_du
        
        self.dx_dv = dx_dv
        self.dy_dv = dy_dv
        self.dz_dv = dz_dv

## Função do Gráfico ##
def grafico(titulo,func, der, color, rot=False):
    X = func.x
    Y = func.y
    Z = func.z
    
    DX_DU = der.dx_du
    DY_DU = der.dy_du
    DZ_DU = der.dz_du
    
    DX_DV = der.dx_dv
    DY_DV = der.dy_dv
    DZ_DV = der.dz_dv

    fig = plt.figure(figsize=(5,5))
    ax = fig.add_subplot(projection = '3d')
    ax.scatter(0, 0, 2, color='r')
    ax.plot_surface(X, Y, Z, rstride = 3, cstride = 3, alpha = 0.8, lw = 0.2, edgecolor='black',color=color,zorder=10)
    ax.plot_surface(DX_DU, DY_DU, DZ_DU, rstride = 3, cstride = 3, lw = 0.2)
    
    ax.set_title(titulo, y = 0.95) 
    plt.rcParams['grid.color'] = "gainsboro"
    #ax.axis('off')
    
    if rot == True:
        rot = np.linspace(0,360,20) # Range da rotação
        caminho = '/home/antenor/you/' # Caminho da pasta onde deseja salvar os frames
        for i in rot:
            ax.view_init(elev = i, azim = i)
            plt.savefig(caminho+str(i)+'.png',dpi=80)
        plt.close()
        


# Função das equações paramétricas #
def parab_elip():
    a = 1
    b = 1
    c = 1
    u = np.linspace(0, 2, 100)
    v = np.linspace(0, 2 * pi, 100)
    u, v = np.meshgrid(u, v)
    x = a * u * np.cos(v)
    y = b * u * np.sin(v)
    z = c * (x**2+y**2)
    return eq_par(x, y, z)

# Função das derivadas parciais # 
### das equações paramétricas ###
def partial_u():
    
    u = np.linspace(-1,1,100)
    v = np.linspace(-1,1,100)
    u, v = np.meshgrid(u, v)
    
    dx_du = u
    dy_du = v
    dz_du = 2 * u
    
    dx_dv = -u * sin(v)
    dy_dv =  u * cos(v)
    dz_dv =  0
    
    return eq_par_der(dx_du, dy_du, dz_du, 
                      dx_dv, dy_dv, dz_dv)
```

## PARABOLÓIDE ELÍPTICO

```python
grafico('Parabolóide Elíptico', parab_elip(), partial_u(),'royalblue')
```

```python
grafico('Parabolóide Elíptico', parab_elip(), partial_u(),'royalblue',rot=True)
```

Equação do hiperbolóide:  
$\Large \frac{x ^ 2}{a ^ 2} \normalsize + \Large \frac{y ^ 2}{b ^ 2} \normalsize = \Large \frac{z}{c}$  
  
  
Equações paramétricas do elipsóide:  
$\normalsize x = a \cdot u \cdot \cos{v}$  
  
$\normalsize y = b \cdot u \cdot \sin{v}$  
  
$\normalsize z = c \cdot u^2$  
  
Com $0 < u < 2$ e $0 < v < 2\pi$.  

Derivadas parciais das equações paramétricas em relação à $u$:

$\Large \frac{\partial x}{\partial u} \normalsize = a \cdot \cos{v}$  
$\Large \frac{\partial y}{\partial u} \normalsize = b \cdot \sin{v}$  
$\Large \frac{\partial z}{\partial u} \normalsize = c \cdot 2 \cdot u$  


Derivadas parciais das equações paramétricas em relação à $v$:

$\Large \frac{\partial x}{\partial u} \normalsize = a \cdot u \cdot \sin{v} \cdot (- 1)$  
$\Large \frac{\partial y}{\partial u} \normalsize = b \cdot u \cdot \cos{v}$  
$\Large \frac{\partial z}{\partial v} \normalsize = 0$
